OpenCv read image in BGR formate so we convert into RGB

```python

new_image = cv2.cvtColor (image , cv2.COLOR_BGR2RGB)
plt.imshow(new_image)
```

Splitting Image Channels

```python

r, g, b = cv2.split(new_image)

# print('r', r.shape)
# print('g', g.shape)
# print('b', b.shape)

new_image = cv2.merge((r,g,b))
```

Resize the Image

```python

s = 10
w = int(new_image.shape[1]*s/100)
h = int(new_image.shape[0]*s/100)
dim=(w,h)
re_size = cv2.resize(new_image,dim,interpolation = cv2.INTER_AREA)
re_size.shape
```

Image Rotation

```python

(h,w) = new_image.shape[:2]

c = (w/2, h/2)

angle = 90

m = cv2.getRotationMatrix2D(c,angle,1.0)
rotate_90 = cv2.warpAffine(new_image,m,(h,w))
```

```python
plt.imshow(rotate_90)
```

Changing Color to Gary and Gray image has only 2 channel

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image_gray.shape
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


plt.figure(figsize=(8, 8))
ax = plt.subplot()
ax.imshow(image_gray, cmap="Greys")
ax.axis('off')
ax.set_title("Gray Image")
plt.show()
```

Resizing and Sacling

```python
image_resized = cv2.resize(image, None, fx=0.25, fy=0.25)
print(image_resized.shape)

ax = plt.subplot()
ax.imshow(image_resized)
ax.axis('off')
ax.set_title("Blur Image")
plt.show()
```

```python
image_resized = cv2.resize(image,(100,200))

ax = plt.subplot()
ax.imshow(image_resized)
ax.axis('off')
ax.set_title("Resized Image")
plt.show()
```

Sharpen Image

```python
kernel_sharpening = np.array([[-1,-1,-1],
                              [-1,9,-1],
                              [-1,-1,-1]])

image_sharpen = cv2.filter2D(image, -1, kernel_sharpening)

ax = plt.subplot()
ax.imshow(image_sharpen)
ax.axis('off')
ax.set_title('Sharpened Image')
plt.show()
```
