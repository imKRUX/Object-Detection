```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread("./maxresdefault.jpg")
```

```python
type(image)
```

```python
image.shape
```

```python
plt.imshow(image)
```

```python
#OpenCv read image in BGR formate so we convert into RGB
new_image = cv2.cvtColor (image , cv2.COLOR_BGR2RGB)
plt.imshow(new_image)
```

```python
"""
Splitting Image Channels

"""
r, g, b = cv2.split(new_image)

# print('r', r.shape)
# print('g', g.shape)
# print('b', b.shape)

new_image = cv2.merge((r,g,b))
```

```python
"""
Resize the Image

"""
s = 10
w = int(new_image.shape[1]*s/100)
h = int(new_image.shape[0]*s/100)
dim=(w,h)
re_size = cv2.resize(new_image,dim,interpolation = cv2.INTER_AREA)
re_size.shape
```

```python
"""
Rotate the Image

"""

(h,w) = new_image.shape[:2]

c = (w/2, h/2)

angle = 90

m = cv2.getRotationMatrix2D(c,angle,1.0)
rotate_90 = cv2.warpAffine(new_image,m,(h,w))
```

```python
plt.imshow(rotate_90)
```
